- `ftcCnclsns`: 공정위 심결례 (해당, 2. 비해당)
- `clauseArticle`: 약관조항
- `dvAntageous`: 유불리판단 (유리, 2. 불리)
- `comProvision`: 비교근거. 공정위 심결례를 받지 않은 유·불리 조항의 판단근거 기입란 : 표준약관 비교조항, 공개 심결례 시정전후 약관 참고 확인 작성
- `illdcssBasiss`: 위법성/판단 근거.
- `relateLaword`: 관련 법령.
- `unfavorableProvision`: 불리한 조항 유형 (신의성실의 원칙 위반/개별금지 조항의 위반)
